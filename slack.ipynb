```python
import numpy as np
import matplotlib.pyplot as plt

# Coordenadas para el landmark 0
x0 = gino[:, 0, 0]
y0 = gino[:, 0, 1]
z0 = gino[:, 0, 2]

# Coordenadas para el landmark 23
x23 = gino[:, 1, 0]
y23 = gino[:, 1, 1]
z23 = gino[:, 1, 2]

# Calcula el tiempo de cada fotograma (asumiendo una tasa de fotogramas constante)
fps = 30  # Ajusta esto segn tu video
time = np.arange(0, len(gino)) / fps

# Grafica las coordenadas x, y, z en funcin del tiempo para ambos landmarks
plt.figure(figsize=(14, 9))

# Coordenadas X
plt.subplot(3, 1, 1)
plt.plot(time, x0, label='Landmark 0 - X')
plt.plot(time, x23, label='Landmark 23 - X', linestyle='--')
plt.ylabel('X coordinate')
plt.legend()

# Coordenadas Y
plt.subplot(3, 1, 2)
plt.plot(time, y0, label='Landmark 0 - Y')
plt.plot(time, y23, label='Landmark 23 - Y', linestyle='--')
plt.ylabel('Y coordinate')
plt.legend()

# Coordenadas Z
plt.subplot(3, 1, 3)
plt.plot(time, z0, label='Landmark 0 - Z')
plt.plot(time, z23, label='Landmark 23 - Z', linestyle='--')
plt.ylabel('Z coordinate')
plt.xlabel('Time (seconds)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_landmarks(gino, landmarks_range, fps=30):
    """
    Plots the X, Y, and Z coordinates of landmarks within the specified range against time.

    Parameters:
    - gino: numpy array containing landmark coordinates.
    - landmarks_range: tuple specifying the range of landmarks to plot (inclusive).
    - fps: frame rate of the video to calculate time.
    """
    time = np.arange(0, len(gino)) / fps  # Calculate time for each frame
    
    plt.figure(figsize=(14, 9))

    # Loop through each coordinate dimension
    for i, coord in enumerate(['X', 'Y', 'Z']):
        plt.subplot(3, 1, i + 1)
        for landmark in range(landmarks_range[0], landmarks_range[1] + 1):
            # Extract coordinates for the current landmark
            x = gino[:, landmark, 0]
            y = gino[:, landmark, 1]
            z = gino[:, landmark, 2]

            if coord == 'X':
                plt.plot(time, x, label=f'Landmark {landmark} - X')
            elif coord == 'Y':
                plt.plot(time, y, label=f'Landmark {landmark} - Y')
            elif coord == 'Z':
                plt.plot(time, z, label=f'Landmark {landmark} - Z')

        plt.ylabel(f'{coord} coordinate')
        plt.xlabel('Time (seconds)')
        plt.legend()
    
    plt.tight_layout()
    plt.show()

# Example usage:
# Assuming `gino` is your numpy array containing the landmarks data
plot_landmarks(gino, (2, 4), fps=30)
```
